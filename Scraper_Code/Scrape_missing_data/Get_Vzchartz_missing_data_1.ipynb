```python
import pandas as pd

import os
```

```python
from scrape_game_user_critic_data import scrape_vg_missing_data
```

```python
progress_tracker = 'Trackers/track_pages.txt'
```

```python
def create_if_not_exists(part_num):
    t_file, ext = progress_tracker.split('.')
    if not os.path.exists(f'{t_file}_{part_num}.{ext}'):
        with open(f'{t_file}_{part_num}.{ext}', 'w') as file:
            file.write('')
```

```python
def record_progress(part, page_num):
    create_if_not_exists(part)
    t_file, ext = progress_tracker.split('.')
    with open(f'{t_file}_{part}.{ext}', 'w') as f:
        f.write(f'{page_num},\n')

def start_page(part):
    create_if_not_exists(part)
    t_file, ext = progress_tracker.split('.')
    with open(f'{t_file}_{part}.{ext}', 'r') as f:
        text = f.read()
    print(text)
    if text == '':
        return 1
    page_num = int(text.split(',\n')[-2].strip())

    return page_num + 1
```

```python
columns = ['Rank','Name',\
           'Critic_Score', 'User_Score'
          ]
```

```python
from math import ceil
```

```python
offset = 1
part = 1
results_each_page = 200

csv_file = f'CSV_Files/vgsales_critic_user_part_{part}.csv'

def scrape_pages(start, end):
    vg_data = pd.read_csv(csv_file) if os.path.exists(csv_file) else None
    for page in range(start, end):
        batch_vg_data = scrape_vg_missing_data(page, results_each_page)
        if vg_data is None:
            vg_data = batch_vg_data
        else:
            vg_data = pd.concat([vg_data, batch_vg_data], ignore_index=True)
        vg_data.to_csv(csv_file, sep=",", encoding='utf-8', index=False)
        record_progress(part, page)
```

```python
results_each_page = 200
total_pages = 105

page_start = max(ceil((total_pages) * offset * (part-1)), start_page(part))
page_end = ceil((total_pages) * offset * (part))
```

```python
print(f'starting at page {page_start} and ending in {page_end}')
scrape_pages(page_start, page_end)
```
